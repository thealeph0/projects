## Taylor Polynomials for Non-Polynomial Functions -- Applied Calculus

If we consider a function $f$ with derivatives of all orders, then the power series generated for $f$ at $x=c$ can be expressed as:
$$f(x) \approx f(c)+\cfrac{f^{(1)}(c)}{1!}(x-c)+ \cfrac{f^{(2)}(c)}{2!}(x-c)^2 +\cdots + \cfrac{f^{(n)}(c)}{n!}(x-c)^n +\cdots  $$

Since this is a long and ongoing sum, we call it an infinite series.  In general mathematics, the uppercase $S$ in Greek is used as an operator for summation: $$ \sum - sigma$$

The Taylor polynomial of order $n$ generated by $f$ at $x=c$ is the polynomial
	$$P_n(x)=f(c)+f'(c)(x-c)+\cfrac{f^{''}(c)}{2!}(x-c)^2 +\cdots + \cfrac{f^{(n)}(c)}{n!}(x-c)^n$$
Note that a Taylor Series is infinite, meanwhile a Taylor Polynomial is of finite degree. Clearly, it is not useful to express functions as infinite sums because we cannot even compute them that way (not even a power computer). However, it is often useful to approximate functions by using a Taylor Polynomial.


```python
import numpy as np
import matplotlib.pyplot as plt
```

In order to plot the sin function along with the first, third, fifth, and seventh order Taylor Polynomial approximations we have provided the following code. Note that this is the zero-th to third in the formula given earlier.

```python
x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

plt.figure(figsize = (10,8))
for n, label in zip(range(4), labels):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    plt.plot(x,y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic Sine') ##actual graph
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

As you can see, the approximation approaches the analytic function quickly, even for $x$
 not near to $a=0$.  Note that in the above code, we also used a new function - 'zip', which can allow us to loop through two parameters range(4) and labels and use that in our plot.
